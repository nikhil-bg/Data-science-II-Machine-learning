```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the CSV file
file_path = '/content/drive/MyDrive/Compozent /CardioGoodFitness.csv'  # Update the path if the file is in a different location
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())

# Summary statistics
print(df.describe())

```

```python
# Pairplot for visualizing relationships
sns.pairplot(df)
plt.show()

```

```python
# Correlation heatmap
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Bar plot for Product distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Product', data=df)
plt.title('Distribution of Products')
plt.show()
```

```python
# Boxplots for numerical variables
plt.figure(figsize=(12, 10))
sns.boxplot(data=df, orient='v')
plt.title('Boxplots of Numerical Variables')
plt.show()
```

```python

# Categorical plots for selected variables
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.barplot(x='Gender', y='Income', data=df)
plt.title('Average Income by Gender')

plt.subplot(2, 2, 2)
sns.boxplot(x='Product', y='Miles', data=df)
plt.title('Miles Run by Product')

plt.subplot(2, 2, 3)
sns.countplot(x='Education', hue='Product', data=df)
plt.title('Product Distribution by Education Level')

plt.subplot(2, 2, 4)
sns.scatterplot(x='Age', y='Miles', hue='Fitness', data=df, palette='viridis')
plt.title('Scatter Plot of Miles vs Age with Fitness Color')

plt.tight_layout()
plt.show()
```
